```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("learningset.csv", low_memory=False)
df
```

### Complete the following steps on the categorical columns in the dataset:


```python
df_cat= df.select_dtypes(object)
df_cat
```

### 1-Check for null values in all the columns


```python
df_cat.isna().sum()
```

### Exclude the following variables by looking at the definitions. Create a new empty list called drop_list. We will append this list and then drop all the columns in this list later:
### SOURCE - symbol definitions not provided, too many categories
### ZIP CODE - we are including state already

```python
drop_list = df_cat.loc[:, ['OSOURCE', 'ZIP']]
drop_list
```

```python
df_cat.drop(drop_list, axis=1, inplace=True) 
df_cat
```

### Identify columns that over 85% missing values



```python
#1st we change the blanks for nan values
df_cat = df_cat.replace(" ", np.nan)
df_cat
```

```python
#2 we p
p85 = len(df_cat)*85/100
p85
```

```python
lost_values = [column for column in df_cat.columns if
           df_cat[column].isna().sum() > p85]
lost_values
```

```python
df_cat[lost_values].isna().sum()
```

### Remove those columns from the dataframe



```python
df_cat.drop(lost_values, axis=1, inplace=True)
```

```python
df_cat.shape
```

### Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest



```python
print(df_cat['GENDER'].value_counts())

df_cat['GENDER'] = df_cat['GENDER'].fillna('F')
```

```python
def new_gender(x):
    if x in ["U", "J", "C", "A"]:
        return "other"
    else: return x
df_cat["GENDER"] = df_cat["GENDER"].apply(lambda x: new_gender(x))
```
